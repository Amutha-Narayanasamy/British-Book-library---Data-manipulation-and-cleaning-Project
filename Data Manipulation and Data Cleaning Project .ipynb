### 3.Check the content of the column- 'Date of Publication’ and define a function to clean the value 

```python
data['Date of Publication']
```

```python
### using regex pattern to clean the values in Date of publication-- for this I am importing the module import re
### patter '\b(\d{4})\b'  is used to match exactly four digits in the date of publication column
```

```python
pattern= r'\b(\d{4})\b'
data['Date of Publication'] = data['Date of Publication'].str.extract(pattern)
```

```python
data
```

```python
## finding the missing values in date of publication column
```

```python
data['Date of Publication'].isnull().sum()
```

```python
data['Date of Publication'].value_counts()
```

```python
## filling with most common value
```

```python
data.dropna(subset=['Date of Publication'],inplace=True)
```

```python
data['Date of Publication'].isnull().sum()
```

### 4. Check the content of the column- ' Author’ and define a function to clean the value. And split the name to first name and last name

```python
## splitting the Author column to First name and last name using split and assignig n= 1 it will split into 2 columns 
## expand function used to create separate column in dataframe.
```

```python
data[['Author_First Name', 'Author_Last Name']] = data['Author'].str.split(n=1, expand=True)
```

```python
data
```

```python
## using rstrip to clean the characters., in the first name column
```

```python
data['Author_First Name']=data['Author_First Name'].str.rstrip('.,')
```

```python
data
```

```python
## using the split function to split the last name column and inside the function assignig '-' to split the text from '-' and expanding the texts 
```

```python
df = data['Author_Last Name'].str.split('-', expand=True)
```

```python
df
```

```python
## Keeping the first column and dropping all other columns in order to get the last name 
```

```python
df.drop(columns=[1,2,3,4,5,6,7,8], inplace=True)
```

```python
df
```

```python
data['Author_Last Name'] = df
```

```python
data
```

```python
## using rstrip to remove the . from last name
## replacing the 'E. (Eneas)','afterwards GILBERT, Davies' 
```

```python
data['Author_Last Name']=data['Author_Last Name'].str.rstrip('.')

data['Author_Last Name']=data['Author_Last Name'].str.replace('E. (Eneas)', 'E')
data['Author_Last Name']=data['Author_Last Name'].str.replace('afterwards GILBERT, Davies', 'Gilbert')
```

```python
data
```

```python
data
```

```python
## droping the Author column using drop function
```

```python
data.drop(columns=['Author'], inplace=True)
```

```python
data
```

```python
data[['Author_First Name', 'Author_Last Name']].isnull().sum()
```

```python
## filling the null values in Author first and last name column with value 'Missing'
```

```python
data['Author_First Name'] = data['Author_First Name'].fillna('Missing')
data['Author_Last Name'] = data['Author_Last Name'].fillna('Missing')
data[['Author_First Name', 'Author_Last Name']].isnull().sum()
```

### 5.Check the content of the column- ‘Title’ and define a function to clean the value

```python
## using regex pattern to capture from the start of the string until '.'
```

```python
pattern = r'^([^.]*)'
data['Title'] = data['Title'].str.extract(pattern)
data
```

```python
## using the lstrip to remove  [ from the start of the string in title column
```

```python
data['Title'] = data['Title'].str.lstrip('[')
data['Title']
```

### 6.Check the content of the column- ‘Place of Publication’ and define a function to clean the value

```python
## Using the regex pattern to extarct the text from place of publication column. 
## This pattern matches any character thai is not a letter or digit or whitespace. It will remove any special characters 
```

```python
data['Place of Publication'] = data['Place of Publication'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
data['Place of Publication'] = data['Place of Publication'].str.split().str[0]
data['Place of Publication']
```

```python
## checking place of publication has null values
```

```python
data['Place of Publication'].isnull().sum()
```

```python
## filling null values with 'unknown'
```

```python
data['Place of Publication']= data['Place of Publication'].fillna('Unknown')
```

```python
data['Place of Publication'].isnull().sum()
```
